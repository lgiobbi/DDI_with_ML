```python
import pandas as pd
```

```python
def set_drug_id_index(df):
    if 'Drug ID' in df.columns:
        return df.set_index('Drug ID')
    elif 'DrugBank ID' in df.columns:
        df.rename(columns={'DrugBank ID': 'Drug ID'}, inplace=True)
        return df.set_index('Drug ID')
    else:
        raise ValueError("No 'Drug ID' or 'DrugBank ID' column found in dataframe.")

def combine_features(feature1_path, feature2_path, output_path):
    # Load the feature files
    feature1_df = set_drug_id_index(pd.read_csv(feature1_path, index_col=0, sep="\t"))
    feature2_df = set_drug_id_index(pd.read_csv(feature2_path, index_col=0, sep="\t"))

    # Print unmatched indices for debugging
    left_unmatched = feature1_df.index.difference(feature2_df.index)
    right_unmatched = feature2_df.index.difference(feature1_df.index)
    print(f"Left unmatched indices: {len(left_unmatched)}")
    print(f"Right unmatched indices: {len(right_unmatched)}")


    combined_df = feature1_df.merge(feature2_df, left_index=True, right_index=True)
    print(f"Combined dataframe shape: {combined_df.shape}")

    df_float = combined_df.select_dtypes(include=['float64', 'float32'])
    display(df_float.head())

    # Save the combined features
    df_float.reset_index(inplace=True)
    df_float.to_csv(output_path, sep="\t")
    return df_float
```

```python
base_path = "/data/giobbi/embeddings/"

feature_1_name = "DESC_GPT"
feature_2_name = "SMILES_GPT"

feature1_path = base_path + feature_1_name + ".csv"
feature2_path = base_path + feature_2_name + ".csv"
output_path = base_path + feature_1_name + "_+_" + feature_2_name + ".csv"
```
