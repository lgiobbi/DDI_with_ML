```python
import pandas as pd
import numpy as np
```

Assume CRESCENDDI and DrugBank have already been cleaned, such that not CRESCENDDI negatives contradict the DrugBank positives. Make sure that graph is treated as undirected in this step, i.e. A-B == B-A.

```python
def sort_row_and_drop_duplicates(df):
    df_sorted = df.copy()
    df_sorted[['Drug1', 'Drug2']] = np.sort(df_sorted[['Drug1', 'Drug2']], axis=1)
    return df_sorted.drop_duplicates()
```

# Combine CRESCENDDI and DrugBank graph

```python
crescenddi_path = "/data/giobbi/CRESCENDDI/CRESCENDDI_wo_contradiction.csv"
drugbank_path = "/data/giobbi/GRAPH/drugbank_graph_wo_contradiction.csv"
```

```python
crescenddi_df = pd.read_csv(crescenddi_path, sep="\t")
drugbank_df = pd.read_csv(drugbank_path, sep="\t")

full_graph = pd.concat([crescenddi_df, drugbank_df])
```

```python
full_graph_sorted = sort_row_and_drop_duplicates(full_graph)

full_graph_sorted.to_csv("/data/giobbi/GRAPH/drugbank_crescenddi_graph_wo_contradiction.csv", sep="\t", index=False)
```
